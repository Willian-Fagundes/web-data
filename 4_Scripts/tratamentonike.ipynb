```python
import numpy as np
import pandas as pd
import re
from unidecode import unidecode
from sqlalchemy import create_engine
```

```python
#Importando dados brutos
df = pd.read_csv('../0_Bases_originais/nikebruto.csv', sep = ";")
df
```

```python
#Drop colunas sem dados
df = df.drop(columns= [" "])
df
```

```python
#nulos
df = df.dropna()
df
```

```python
#Drop duplicados
df = df.drop_duplicates()
df
```

```python
df['Tipo'] = df['Nome'].str.split().str[0]
```

```python
df
```

```python
#resetar o index
df2 = df.reset_index(drop=True)
```

```python
#Unidecode
df2["Nome"] = df2["Nome"].apply(lambda x: re.sub(r'[^a-zA-Z0-9]','', unidecode (x)))
df2["Porcentagem desc"] = df2["Porcentagem desc"].apply(lambda x: re.sub(r'[^a-zA-Z0-9]','', unidecode (x)))
df2["Tipo"] = df2["Tipo"].apply(lambda x: re.sub(r'[^a-zA-Z0-9]','', unidecode (x)))
df2
```

```python
#Retirar strings 
df2["Valor c desc"] = df2["Valor c desc"].str.replace("R$", "").str.replace("no Pix", "").str.replace(",", "")
df2["Valor s desc"] = df2["Valor s desc"].str.replace("R$", "").str.replace(",", "")
df2["Porcentagem desc"] = df2["Porcentagem desc"].str.replace("off", "")
df2 = df2[~df2['Nome'].str.contains('Personalize', na=False)]
df2 = df2[~df2['Avaliacao'].str.contains('Disponível em SNKRS', na=False)]
df2
```

```python
#transformar strings em valores numericos
df2 ["Porcentagem desc"] = df2 ["Porcentagem desc"].astype(float)
df2["Valor s desc"] = df2["Valor s desc"].astype(float)
df2["Valor c desc"] = df2["Valor c desc"].astype(float)
df2["Avaliacao"] = pd.to_numeric(df2["Avaliacao"])
df2
```

```python
#Divisão para os valores float não ficarem altos demais
df2["Valor c desc"] = df2["Valor c desc"]/100
df2["Valor s desc"] = df2["Valor s desc"]/100
df2
```

```python
#Reset index de novo?
df3 = df2.reset_index(drop=True)
df3
```

```python
# Descobrir outliers
valorQ1 = df3['Valor c desc'].quantile(0.25)
valorQ3 = df3['Valor c desc'].quantile(0.75)
IQRvalor = valorQ3 - valorQ1

# Identificando os limites para os outliers
valor_minimo = valorQ1 - 1.5 * IQRvalor
valor_maximo = valorQ3 + 1.5 * IQRvalor

# Filtrando os outliers
outliers_valor = df3[(df3['Valor c desc'] < valor_minimo) | (df3['Valor c desc'] > valor_maximo)]

outliers_valor
```

```python
descQ1 = df3["Porcentagem desc"].quantile(0.25)
descQ2 = df3["Porcentagem desc"].quantile(0.75)
IQRdesc = descQ2 - descQ1

desc_minimo = descQ1 - 1.5 * IQRdesc
desc_maximo = descQ2 + 1.5 * IQRdesc

outlier_desc = df3[(df3["Porcentagem desc"] < desc_minimo)|(df3["Porcentagem desc"] > desc_maximo)]

print (outlier_desc)
```

```python
avQ1 = df3["Avaliacao"].quantile(0.25)
avQ2 = df3["Avaliacao"].quantile(0.75)
IQRav = avQ2 - avQ1

avaliacao_minimo = avQ1 - 1.5 * IQRav
avaliacao_maximo = avQ2 + 1.5 * IQRav

outliers_avaliacao = df3[(df3["Avaliacao"] < avaliacao_minimo) | (df3["Avaliacao"] > avaliacao_maximo)]
print (outliers_avaliacao)
```

```python

df4 = df3[(df3['Valor c desc'] >= valor_minimo) & (df3['Valor c desc'] <= valor_maximo)]
df4
```

```python
#Ultimo reset index
dffinal = df4.reset_index(drop= True)
```
